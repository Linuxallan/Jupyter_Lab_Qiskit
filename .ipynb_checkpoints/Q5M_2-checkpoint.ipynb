```python
# He comprendido el cx 'CNOT'
### cuando el qubit desde donde empieza el CNOT esta en 1 
# lo que hace es ver el valor del cubit al cual esta dirigido el CNOT y hace un XOR
# XOR : si ambos son iguales en valor : el resultado se mantiene
# --> : si ambos valores son opuestos (1 y 0, o 0 y 1) : el estado del Qubit
# al que va dirigido el CNOT cambia al opuesto.
# si el valor a voltear es 50/50 -> se mantiene en 50/50
### Si el qubit de donde empieza el CNOT esta en valor 0
# deja intacto el estado del qubit al que esta dirigido el CNOT
### Si el qubit de donde empieza el CNOT esta en valor 50/50
# lo que hace es sobreescribir el estado del qubit destino en valor 50/50
```

```python
from qiskit import *
from qiskit.tools.visualization import plot_histogram
from matplotlib import *
```

```python
circuit = QuantumCircuit(3,3)
circuit.x(0)
#circuit.h(0)
circuit.barrier()
```

```python
circuit.draw(output='mpl')
```

```python
circuit.h(1)
circuit.cx(1,2)
circuit.cx(0,1)
circuit.h(0)
circuit.barrier()
```

```python
circuit.draw(output='mpl')
```

```python
#circuit.measure([0,1],[0,1])
circuit.barrier()
```

```python
circuit.draw(output='mpl')
```

```python
circuit.cx(1,2)
circuit.cx(0,2)
```

```python
circuit.draw(output='mpl')
```

```python
circuit.barrier()
circuit.measure(2,2)
```

```python
circuit.draw(output='mpl')
```

```python
simulator = Aer.get_backend('qasm_simulator')
result = execute(circuit, backend = simulator, shots = 1024).result()
counts = result.get_counts()
plot_histogram(counts)
```

```python
print(counts)
```

```python
circ = QuantumCircuit(6,6)
circ.x(0)
circ.cx(0,1)
circ.measure(1,1)
```

```python
circ.h(2)
circ.x(3)
circ.cx(2,3)
circ.measure(3,3)
circ.draw(output='mpl')
```

```python
print(coun)
```

```python
circ.x(4)
circ.h(5)
circ.cx(4,5)
circ.measure(5,5)
circ.draw(output='mpl')
```

```python
# q0 -> 1
# q1 -> 0  ::-> 1

# q2 -> H
# q3 -> 1  ::> 50/50

# q4 -> 1
# q5 -> H  :: 50/50
```

```python
result = execute(circ, backend = simulator, shots = 1024).result()
con = result.get_counts()
print(con)
```

```python
plot_histogram(con)
```
