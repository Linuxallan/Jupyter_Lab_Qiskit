```python
from qiskit import *
from qiskit.tools. visualization import plot_histogram
```

```python
n = 8
```

```python
circuit = QuantumCircuit(n,n)
```

```python
for i in range(8):
    circuit.measure(i,i)
circuit.draw(output='mpl')
```

```python
simulator = Aer.get_backend('qasm_simulator')
# 'deqcaobj' : significa -> de quantum circuit a objeto
# 'assemble' transforma el circuito cuantico a objeto
deqcaobj = assemble(circuit)
result = simulator.run(deqcaobj).result()
counts = result.get_counts()
print(counts)
```

```python
#
# Sumar circuitos cuanticos
#
```

```python
circuit_2 = QuantumCircuit(n)
circuit_2.x(3)
circuit_2.x(4)
circuit_2.barrier()
# circuito 3
circuit_3 = circuit_2 + circuit
circuit_3.draw(output='mpl')
```

```python
obj = assemble(circuit_3)
counts = simulator.run(obj).result().get_counts()
print(counts)
```

```python
circuit = QuantumCircuit(2,2)
circuit.x(0)
#circuit.x(1)
circuit.cx(0,1)
circuit.measure(0,0)
circuit.measure(1,1)
circuit .draw(output='mpl')
```

```python
circuit = QuantumCircuit(3,1)
circuit.x(0)
circuit.cx(0,1)
circuit.ccx(0,1,2)
circuit.measure(2,0)
circuit.draw(output='mpl')
```

```python
counts = execute(circuit, backend = simulator, shots = 1024).result().get_counts()
print(counts)
```
