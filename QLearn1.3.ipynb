```python
from qiskit import *
from qiskit.tools.visualization import *
from math import sqrt, pi
```

```python
qc = QuantumCircuit(1)
# inicializar el qubit en estado 0
initial_state = [0,1]
# inicializar el qubit en super posicion
#initial_state = [0.+1.j/sqrt(2), 1/sqrt(2)+0.j]
#
# initialize(initial_state(estado del cubit), 0(el numero del qubit que quiere
# iniciar en ese estado) )
qc.initialize(initial_state, 0)
qc.draw(output='mpl')
```

```python
# 'statevector' permite ver el estado del qubit antes de medirlo
# pero es solo aducativo, en realidad es algo que no se puede
# El 'qasm_simulator' no permite ver el estado de los qubit antes de medirlo
# por que el 'qasm_simulator' se comporta como una Quantum computing real.
simulvector = Aer.get_backend('statevector_simulator')
obj = assemble(qc)
result = simulvector.run(obj).result()
#counts = result.get_counts()
out_state = result.get_statevector()
#print(counts)
print(out_state)
# Python usa 'j' para representar 'i'(imaginario) en numero complejos
```

```python
qc.measure_all()
qc.draw(output='mpl')
```

```python
obj = assemble(qc)
result = simulvector.run(obj).result()
counts = result.get_counts()
print(counts)
```

```python
plot_histogram(counts)
```
